```python
#padronizaao
df_six = df.copy()
colunas_numericas = df_six.select_dtypes(include = 'number').columns
df_six[colunas_numericas] = pd.DataFrame(data=StandardScaler().fit_transform(df_six[colunas_numericas]), 
                                         columns=df_six[colunas_numericas].columns)
df_six
```

```python
columns_to_drop = df_six.select_dtypes(include=['object', 'string']).columns
df_six = df_six.drop(columns=columns_to_drop)
df_six.head()
```

## Nmero de grupos

Utilize as tcnicas vistas em aula que te ajudem a decidir pelo nmero de grupos a ser utilizados.

```python
sse = []
k_range = range(1, 11)  
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42).fit(df_six)
    sse.append(kmeans.inertia_)  
```

```python
plt.plot(k_range, sse, 'bx-')  
plt.xlabel('Nmero de Clusters (k)')
plt.ylabel('Soma dos Erros ao Quadrado (SSE)')
plt.title('Tcnica do Cotovelo para K-means')
plt.show()
```

```python
silhuetas = []

max_clusters = 10

for n_clusters in tqdm(range(2, max_clusters + 1)):
    
    km = KMeans(n_clusters = n_clusters, 
                n_init = 'auto').fit(df_six[colunas])
    
    silhuetas.append(silhouette_score(X = df_six[colunas], 
                                      labels = km.labels_))

    nomes_grupos = ['grupo_' + str(n) for n in range(n_clusters)]
    
    df_six['grupos_'+str(n_clusters)] = pd.Categorical.from_codes(codes = km.labels_, 
                                                                  categories = nomes_grupos)    

df_silhueta = pd.DataFrame(data = {'n_clusters': list(range(2, max_clusters + 1)), 
                                   'silhueta_media': silhuetas})
print(df_silhueta)
df_silhueta.plot.line(x = 'n_clusters', 
                      y = 'silhueta_media', 
                      marker = 'o')

plt.show()
```

## Avaliao dos grupos

Faa uma anlise descritiva para pelo menos duas solues de agrupamentos (duas quantidades diferentes de grupos) sugeridas no item anterior, utilizando as variveis que esto no escopo do agrupamento.
- Com base nesta anlise e nas anlises anteriores, decida pelo agrupamento final. 
- Se puder, sugira nomes para os grupos.

```python
def cluster_barplot(data: pd.DataFrame, coluna_cluster: str) -> None:
  
    df_agrupado = data.groupby([coluna_cluster, 'Revenue'])['Revenue'].count().unstack()
    
    ax = df_agrupado.plot.bar(rot=0)
    
    ax.set_ylabel('Quantidade de clientes')
    ax.set_xlabel('Clusters')
    plt.title('Quantidade de clientes por cluster')
    
    plt.show()

cluster_barplot(df_six, 'grupos_5')    

cluster_barplot(df_six, 'grupos_10')
```

```python
df_six_ = df_six.copy()

df_six_['grupos_2'] = df_six_['grupos_2'].cat.rename_categories({'grupo_0': 'No interessados', 
                                                                 'grupo_1': 'Compradores'})

df_six_['grupos_3'] = df_six_['grupos_3'].cat.rename_categories({'grupo_0': 'No interessados', 
                                                                 'grupo_1': 'Potenciais compradores', 
                                                                 'grupo_2': 'Compradores'})
```

```python
cluster_barplot(df_six_, 'grupos_2')

cluster_barplot(df_six_, 'grupos_3')
```

```python
for n in range(6):
    fig, ax = plt.subplots(1, 2, figsize=(16, 4))
    sns.boxplot(data=df_six_, x=colunas[n], y='grupos_2', ax=ax[0])
    sns.boxplot(data=df_six_, x=colunas[n], y='grupos_4', ax=ax[1])
    plt.subplots_adjust(wspace=0.5)

    plt.show()
```

## Avaliao de resultados

Avalie os grupos obtidos com relao s variveis fora do escopo da anlise (minimamente *bounce rate* e *revenue*). 
- Qual grupo possui clientes mais propensos  compra?

```python
pd.crosstab(df_six_['Revenue'], df_six_['grupos_2'], normalize='all', margins=True)
```

<blockquote>
No caso de um agrupamento com 2 clusters, podemos nomear o grupo_0 como "No interessados", referindo-se a clientes com menor chance de realizar compras. J o grupo_1 pode ser chamado de "Compradores", representando aqueles com maior probabilidade de comprar.

Embora haja uma diferena significativa na proporo de clientes marcados como "False" e "True" na varivel 'Revenue', o grupo_0 conseguiu identificar a maioria dos clientes com menor propenso a comprar. Curiosamente, este grupo acertou cerca de 74% dos casos onde a varivel 'Revenue'  "False" (indicando que os clientes no compraram), o que representa 85% do total de registros. Por outro lado, o grupo_1, que engloba os clientes mais propensos a comprar, acertou aproximadamente 4% dos casos onde a varivel 'Revenue'  "True" (indicando compradores), correspondendo a 15% do total de dados.</blockquote>
