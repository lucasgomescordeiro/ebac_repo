# Probabilidade - Tarefa

```python
import math
import numpy             as np
import pandas            as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn           as sns

from scipy import stats
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
media = 15 
desvio_padrao = 0.1 

limite_superior = 15.2

probabilidade = 1 - stats.norm.cdf(x=limite_superior, loc=media, scale=desvio_padrao)

print(f'A probabilidade de um pacote ter mais que {limite_superior} kg  de: {probabilidade*100:.2f}%'.replace('.', ','))
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
percentil = stats.norm.ppf(q=0.95, loc=media, scale=desvio_padrao)

print(f'O percentil de ordem 95% para o peso dos pacotes : {percentil:.2f} kg'.replace('.', ','))
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
Y = np.linspace(-6, 6, 200)

media = 0
variancia = 1
desvio_padrao = math.sqrt(variancia)

Y1 = stats.norm.cdf(x=Y, loc=media, scale=desvio_padrao)

graus_de_liberdade = 5

Y2 = stats.t.cdf(x=Y, df=graus_de_liberdade)

# Exibe os resultados em um DataFrame
pd.set_option('display.max_rows', 200)
pd.DataFrame(data={'Y':map(lambda x: f'<= {x:.2f}', Y), 
                   'Y1':map(lambda x: f'{round(x*100, 2)}%', Y1), 
                   'Y2':map(lambda x: f'{round(x*100, 2)}%', Y2)})
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
sns.set(style='whitegrid')

fig, ax = plt.subplots(figsize=(30, 9))

ax.plot(Y, Y1, label='Y1 - Normal')
ax.plot(Y, Y2, label='Y2 - t(5)')

ax.xaxis.set_major_locator(plt.MaxNLocator(14))
ax.yaxis.set_major_locator(plt.MaxNLocator(11))

ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1))

ax.set_xlabel('Valores de Y')

ax.set_ylabel('Probabilidade Acumulada')

ax.legend()

plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
P = np.linspace(0, 1, 20)
# Seu cdigo comea aqui

media = 0
variancia = 1
desvio_padrao = math.sqrt(variancia)

Y1 = stats.norm.ppf(q=P, loc=media, scale=desvio_padrao)

graus_de_liberdade = 5

Y2 = stats.t.ppf(q=P, df=graus_de_liberdade)

pd.DataFrame(data={'P':map(lambda x: round(x, 2), P), 
                   'Y1':map(lambda x: round(x, 2), Y1), 
                   'Y2':map(lambda x: round(x, 2), Y2)})
```

6. Faa um grfico com os dados do item anterior.

```python
fig, ax = plt.subplots(figsize=(30, 9))

ax.plot(P, Y1, label='Y1 - Normal')
ax.plot(P, Y2, label='Y2 - t(5)')

ax.xaxis.set_major_locator(plt.MaxNLocator(20))
ax.yaxis.set_major_locator(plt.MaxNLocator(22))

ax.set_xlabel('Percentis (P)')

ax.set_ylabel('Valores de Y1 e Y2')

ax.legend()

plt.show()
```
