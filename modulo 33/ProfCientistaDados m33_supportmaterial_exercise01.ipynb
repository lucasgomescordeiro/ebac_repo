## 1) Diferena de peso entre machos e fmeas

Na base de pinguins, podemos ver que h diferena de peso entre machos e fmeas. Mas ser que  uma concluso extrapolvel a toda a populao de pinguins? Assumindo que essa amostra  aleatria e representativa da populao de pinguins, realize um teste T com 5% de significncia para verificar se podemos induzir esta concluso para a populao toda.

```python
# Definio das hipteses nula e alternativa
print('''
Hiptese nula       H0: Peso dos machos <= Peso das fmeas
Hiptese alterativa H1: Peso dos machos >  Peso das fmeas''')

# Machos
male = pg.loc[pg.sex == 'Male', 'body_mass_g'] # Amostras de peso dos machos
n_male = len(male) # Tamanho das amostras
mean_male = np.mean(male) # Mdia
std_male = np.std(male, ddof=1) # Desvio padro

# Fmeas
female = pg.loc[pg.sex == 'Female', 'body_mass_g'] # Amostras de peso das fmeas
n_female = len(female) # Tamanho das amostras
mean_female = np.mean(female) # Mdia
std_female = np.std(female, ddof=1) # Desvio padro

# Diferena entre as mdias
mean_difference = mean_male - mean_female

# Erro padro
std_error = np.sqrt((std_male**2/n_male) + (std_female**2/n_female))

# Estatstica t
t_statistic = mean_difference / std_error

# Graus de liberdade
df = n_male+n_female-2

# Valor crtico do teste t para um intervalo de confiana de 95%
alpha = 0.05
t_critical = stats.t.ppf(q=1-alpha, df=df, loc=0, scale=1) # q=95%

# Valor-p para o teste t unilateral  direita
p_value = 1-stats.t.cdf(x=t_statistic, df=df)

# Apresentao dos resultados
print(f'''
Resultados:

Diferena entre as mdias: {mean_difference:.2f}
Desvio padro (Macho): {std_male:.2f}
Desvio padro (Fmea): {std_female:.2f}
Erro padro: {std_error:.2f}
Estatstica do teste: {t_statistic:.2f}
Graus de liberdade: {df}
Regio crtica: {t_critical:.2f}
P-value: {p_value} ({p_value:.18f})
''')

# Interpretao dos resultados
if t_statistic > t_critical:
    print("Estatstica do teste > Regio crtica")
else:
    print("Estatstica do teste < Regio crtica")
    
if p_value < 0.05:
    print("P-value < 5% \n\n Rejeitar a hiptese nula. H evidncias de que o peso dos machos  significativamente maior que o peso das fmeas.")
else:
    print("P-value > 5% \n\n No h evidncias suficientes para rejeitar a hiptese nula. No h diferena significativa no peso entre machos e fmeas.")
```

```python
x1 = pg.loc[pg['sex'] == 'Male', 'body_mass_g']
x2 = pg.loc[pg['sex'] == 'Female', 'body_mass_g']

# Teste t
t_statistic, p_value, degrees_of_freedom = ttest_ind(x1=x1, x2=x2, alternative='larger', usevar='unequal', value=0)

# Resultados
results = {
    'test statistic': t_statistic,
    'pvalue of the t-test': p_value,
    'degrees of freedom used in the t-test': degrees_of_freedom
}

# Apresentao dos resultados
print("Resultados:\n")
for label, value in results.items():
    print(f"{label}: {value:.18f}")

# Interpretao dos resultados
if p_value < 0.05:
    print("\nRejeitar a hiptese nula. H evidncias de que o peso dos machos  significativamente maior que o peso das fmeas.")
else:
    print("\nNo h evidncias suficientes para rejeitar a hiptese nula. No h diferena significativa no peso entre machos e fmeas.")
```

## 2) (opcional) Representao grfica do teste

Siga os passos da aula para voc poder visualizar a estatstica do teste, sua distribuio e a regio crtica do teste.

```python
# Representao grfica do teste
fig, ax = plt.subplots(figsize=(24, 6))

# Parmetros da densidade sob a hiptese nula
mu = 0 # mdia
variance = 1 # varincia
sigma = variance**.5 # desvio padro

# Valores de X e Y para plotar a densidade da distribuio t
x = np.linspace(np.floor(mu-(t_statistic*sigma)), np.ceil(mu+(t_statistic*sigma)), 100000)
y = stats.t.pdf(x, df)

# Grfico da densidade
ax.plot(x, y, label='Densidade t')

# Regio Crtica
low = stats.t.ppf(.95, df, loc=0, scale=1)
high = np.inf

# rea aceitvel
ax.fill_between(x, y, where=(x<low), alpha=0.25, label='rea Aceitvel', color='green')

# rea da regio crtica
ax.fill_between(x, y, where=(x>low)&(x<high), alpha=0.25, label='Regio Crtica', color='red')

# Linha vertical para o ponto crtico da linha crtica
ax.vlines(x=t_critical, ymin=y.min(), ymax=stats.t.pdf(t_critical, df), color='r', linestyle='--', label='Ponto de Corte da Linha Crtica')

# Linha vertical para a estatstica do teste
ax.vlines(x=t_statistic, ymin=y.min(), ymax=y.max(), color='black', linestyle='dotted', label='Estatstica do Teste')

# Ajustar a quantidade de marcas no eixo x
ax.set_xticks(np.arange(x.min(), x.max()+.5, .5))

# Legenda e rtulos
ax.legend(loc='upper left')
ax.set_title('Distribuio t e Teste de Hiptese')
ax.set_xlabel('Valor da Estatstica do Teste')
ax.set_ylabel('Densidade da Distribuio t')

plt.show()

# Probabilidade de X pertencer  rea
prob_area = stats.t.cdf(high, df) - stats.t.cdf(low, df)
print(f'A probabilidade para valores entre {low:.2f} e {high:.2f}  de {prob_area:.2%}'.replace('.',','))
```

## 3) Considere as espcies

Repita o exerccio I, uma vez para cada espcie de pinguim. Qual  o *p-value* associado a cada raa?

```python
for specie in pg.species.unique():
    # Species
    print(f'\tEspcie: {specie}')
    
    # Definio das hipteses nula e alternativa
    print('''
    Hiptese nula       H0: Peso dos machos <= Peso das fmeas
    Hiptese alterativa H1: Peso dos machos >  Peso das fmeas''')

    # Machos por espcie
    male = pg.loc[(pg.sex == 'Male') & (pg.species == specie), 'body_mass_g']
    n_male = len(male) # Tamanho das amostras
    mean_male = np.mean(male) # Mdia
    std_male = np.std(male, ddof=1) # Desvio padro

    # Fmeas por espcie
    female = pg.loc[(pg.sex == 'Female') & (pg.species == specie), 'body_mass_g']
    n_female = len(female) # Tamanho das amostras
    mean_female = np.mean(female) # Mdia
    std_female = np.std(female, ddof=1) # Desvio padro

    # Diferena entre as mdias
    mean_difference = mean_male - mean_female
    
    # Erro padro
    std_error = np.sqrt((std_male**2/n_male) + (std_female**2/n_female))
    
    # Estatstica t
    t_statistic = mean_difference / std_error

    # Graus de liberdade
    df = n_male+n_female-2

    # Valor crtico do teste t para um intervalo de confiana de 95%
    alpha = 0.05
    t_critical = stats.t.ppf(q=1-alpha, df=df) # q=95%

    # Valor-p para o teste t unilateral  direita
    p_value = 1-stats.t.cdf(x=t_statistic, df=df)

    # Apresentao dos resultados
    print(f'''
    Resultados:

    Diferena entre as mdias: {mean_difference:.2f}
    Desvio padro (Male): {std_male:.2f}
    Desvio padro (Female): {std_female:.2f}
    Erro padro: {std_error:.2f}
    Estatstica do teste: {t_statistic:.2f}
    Graus de liberdade: {df}
    Regio crtica: {t_critical:.2f}
    P-value: {p_value} ({p_value:.18f})
    ''')

    # Interpretao dos resultados
    if t_statistic > t_critical:
        print("Estatstica do teste > Regio crtica")
    else:
        print("Estatstica do teste < Regio crtica")

    if p_value < 0.05:
        print(f"P-value < 5% \n\n Rejeitar a hiptese nula. H evidncias de que o peso dos machos da espcie {specie}  significativamente maior que o peso das fmeas.\n")
    else:
        print(f"P-value > 5% \n\n No h evidncias suficientes para rejeitar a hiptese nula. No h diferena significativa no peso entre machos e fmeas da espcie {specie}.\n")
```

```python
for specie in pg.species.unique():
    # Dados por espcie
    x1 = pg.loc[(pg['species'] == specie) & (pg['sex'] == 'Male'), 'body_mass_g']
    x2 = pg.loc[(pg['species'] == specie) & (pg['sex'] == 'Female'), 'body_mass_g']

    # Teste t
    t_statistic, p_value, degrees_of_freedom = ttest_ind(x1=x1, x2=x2, alternative='larger', usevar='unequal', value=0)

    # Resultados
    results = {
        'test statistic': t_statistic,
        'pvalue of the t-test': p_value,
        'degrees of freedom used in the t-test': degrees_of_freedom
    }

    # Apresentao dos resultados
    print(f"\tResultados para a espcie {specie}:\n")
    for label, value in results.items():
        print(f"{label}: {value:.18f}")

    # Interpretao dos resultados
    if p_value < 0.05:
        print("Rejeitar a hiptese nula. H evidncias de que o peso dos machos  significativamente maior que o peso das fmeas.\n")
    else:
        print("No h evidncias suficientes para rejeitar a hiptese nula. No h diferena significativa no peso entre machos e fmeas.\n")
```
