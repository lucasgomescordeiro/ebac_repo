### Frmula de bhaskara

<p> A soluo para uma funo de segundo grau depende das suas raises (valores de X). Como j vimos , os coeficientes <br>
    precisa  ser numeros reais e o angular diferente de zero. Sendo assim, temos a seguinte frmula de bhaskara:</p>
    <br>
    
    
   $$ x=\frac{-b\frac{+}{}\sqrt{b^2-4.a.c}}{2.a} $$
   
   <br>
   <center>Frmula de bhaskara.</center>
   <br>
   <br>
<p> Para melhor compreenso o seu calculo  dividido em duas partes: discriminante da equao e operaes para <br>
    determinar as raizes.</p>
   
####  Discriminante

<p> A figura dentro da raiz da frmula de bhaskara  nomeada de discriminante. Seu simbolo  a letra grega delta e<br>
    apresenta a determinada formula:
    <br>
    
   $$ \Delta = b^2-4.a.c $$
    <br>
    <center>Frmula da discriminante.</center>
    <br>
- Se delta for maior que zero, a equao ter dois valores reais distintos.
- Se o delta for igual a zero, a equao tera somente um valor real ou dois resultados iguais.
- Se o delta for menor que zero, a equao no possuir valores reais.
